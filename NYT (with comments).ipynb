```python
from datetime import datetime

#open the bestseller file
bestsellers = open('bestsellers.txt')
#create a list and add all the books in it
all_list = []
for line in bestsellers:
    line_list = line.split('\t')
    all_list.append(line_list)

#the interface that is displayed at the opening of the program
def interface1():
    print("Welcome to the program! What do you want to do?")
#add an empty line for a more convenient reading
    print("\n")
    print("1: Search the bestseller books in a year range.")
    print("2: Search for all books in a specific month and year.")
    print("3: Search for an author.")
    print("4: Search for a title.")
    print("5: My Birthday book!")
    print("Q: Quit the application.")

#the interface that is displayed each time after the first one
def interface2():
    print("What do you want to do next?")
#add an empty line for a more convenient reading
    print("\n")
    print("1: Search the bestseller books in a year range.")
    print("2: Search for all books in a specific month and year.")
    print("3: Search for an author.")
    print("4: Search for a title.")
    print("5: My Birthday book!")
    print("Q: Quit the application.")

#creation of a new function that search for a year range in the list
def search_years():
    try:
#asks the user for the range he wants to search within     
        years_start = input("Enter a starting year:")
        years_end = input("Enter an ending year:")
#add an empty line for a more convenient reading
        print("\n")

#creates a list with all years that are in this range 
        years_lst = []
        for i in range(int(years_start), int(years_end), 1):
            years_lst.append(i)
        years_lst.append(years_end)

#As the list is made of integers, we convert them into strings
        years_lst_in_string = [str(item) for item in years_lst]

#checking that the user entered an ending year that is higher than the starting year    
        if years_end < years_start: 
            print("The starting year should be smaller than the ending year\n")

#If the user entered the input correctly, we continue the function         
        else:  
#looks for all books that contains one of the years in its line on the list and display them
            print("Here is a list of all books that reached #1 top seller book between", years_start, "and", years_end, ":\n")
            for book in all_list:
                for i in years_lst_in_string:
                    if i in book[3]:
                        print(book[0],", by", book[1], "(" , book[3],")\n")
            
  #if the user did not enter a number for the years, an error message is displayed
    except ValueError:
        print("This is not a valid year.\n")

#definition of a new function that looks for a specific month and year
def search_monthyear():
    try:
#ask the user for the month and year he wants to search
        m = int(input("Enter month (as number, 1-12):"))
        y = int(input("Enter a year:"))
#add an empty line for a more convenient reading
        print("\n")
#create a condition for month being max 12 and year being in the range that can be found in the bestseller list 
        if m<=12 and 1942<=y<=2013:
            print("All titles in month", m, "of", y, ":\n" )
#search all books that contain the month and year specified and display them
            for book in all_list:
                objDate = datetime.strptime(book[3], '%m/%d/%Y')
                if objDate.year == y and objDate.month == m:
                    print(book[0],", by", book[1], "(" , book[3],")\n")
            else:
                print("Sorry, there is no bestseller in this given month and year. \n")
#if the user did not enter a correct month and/or year, there is an error message
        else:
            print("This year and/or month does not figure in our database.\n")

#if the user did not enter an integer for month or year there is an error message        
    except ValueError: 
        print("There is something wrong with the month or the year. \n")

#creation of a new function that looks up for book titles
def search_authors():
    lst =[]
#ask for the author the user wants to search
    author = input("What is the name of the author you are looking for?")
    author = ' '.join(word[0].upper() + word[1:].lower() for word in author.split())
#add an empty line for a more convenient reading
    print("\n")
    print("Here is a list of the books containing", author, "as author:\n")
#display all books that contain the user's input as part of the author's name
    for book in all_list:
        if author in book[1]:
            print(book[0],", by", book[1], "(" , book[3],")\n")
            continue
        else: lst.append(book)
#if there is no match with the user's input, this will be displayed
        if lst == all_list: 
            return print ('{} is not a bestseller author.'.format(author))

#creation of a new function that looks for the authors
def search_title():
    lst=[]
#ask the user for the title he wants to search
    title = input("What is the title you are looking for?")
#add an empty line for a more convenient reading
    print("\n")
    title = ' '.join(word[0].upper() + word[1:].lower() for word in title.split()) 
    print("Here is a list of the books with the word(s)", title, "in it:\n")
#display all books that contain the user's input as part of the title
    for book in all_list:
        if title in book[0]:
            print(book[0],", by", book[1], "(" , book[3],")\n")
            continue
        else: lst.append(book)
#if there is no book with the user's input, this will be displayed
    if lst == all_list: 
        return print ('{} is not a bestseller title.'.format(title))

#creation of a function that look up if a bestseller reached #1 on the user's birthday or any other specific date
def birthday():
    lst = [] 
    try:
#ask the user for its day, month and year of birthday
        birth = input('Your month (1-12):'), input('Your day: '), input('Your year: ')
        birthday = "/".join(birth)
#add an empty line for a more convenient reading
        print("\n")
#look for all books in the list if one has reached #1 on the date entered and display it if it is the case
        for book in all_list:
            if birthday in book[3]:
                print(book[0]," by", book[1], "became #1 top seller on your birthday (", book[3], ")\n")
                continue
            else:
                lst.append(book)
  #if no book reached #1 on the day entered, this will be displayed    
        if lst == all_list:
            return print('Sorry, no book became #1 best seller on your birthday.\n')

    except ValueError:
        print("This is not a valid choice, please try again!")

choice = ()
#create a count so that this interface is only displayed the first time
count = 0 
while count == 0:
    interface1()
#add an empty line for a more convenient reading
    print("\n")
#link all the functions created depending on the user's choice
    choice = input("Please enter your choice:")
#add an empty line for a more convenient reading
    print("\n")
    if choice == "1":
        search_years()
        count += 1 
    elif choice =="2": 
        search_monthyear()
        count += 1 
    elif choice =="3":
        search_authors()
        count += 1 
    elif choice =="4":
        search_title()
        count += 1 
    elif choice =="5":
        birthday()
        count+= 1
    elif choice.lower() == 'q':
#add an empty line for a more convenient reading
        print("\n")
        print("Thank you for using the best program ever!") 
        break
#if the user is not entering a valid choice, an error message is displayed
    else:
#add an empty line for a more convenient reading
        print("\n")
        print("This is not a valid choice, please try again!")
#add an empty line for a more convenient reading
        print("\n")
        count += 1 

#display the second interface 
while choice != "q":
    interface2()
#add an empty line for a more convenient reading
    print("\n")
#link all the functions created depending on the user's choice
    choice = input("Please enter your choice:")
    if choice == "1":
        search_years()
    elif choice =="2": 
        search_monthyear()
    elif choice =="3":
        search_authors()
    elif choice =="4":
        search_title()
    elif choice =="5":
        birthday()
    elif choice.lower() == 'q':
#add an empty line for a more convenient reading
        print("\n")
        print("Thank you for using the best program ever!") 
        break
#if the user is not entering a valid choice, an error message is displayed
    else:
#add an empty line for a more convenient reading
        print("\n")
        print("This is not a valid choice, please try again!")
#add an empty line for a more convenient reading
        print("\n")
```
